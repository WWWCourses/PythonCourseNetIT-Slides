```python
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib: 3D Ploting 

3D plotting is particularly useful when visualizing complex datasets or understanding relationships between three variables. 

A three-dimensional axes can be created by passing the keyword `projection='3d'` to any of the normal axes creation routines.

Note that prior to Matplotlib 3.2.0, it was necessary to explicitly import the mpl_toolkits.mplot3d module to make the '3d' projection to Figure.add_subplot.

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
```

## Basic 3D plot

Let's begin with a basic 3D plot. We'll plot a line connecting three points in 3D space:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define data
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

# Plot the line
ax.plot(x, y, z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Basic 3D Line Plot')

# Show the plot
plt.show()
```

## Scatter plot in 3D

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate random data
np.random.seed(0)
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Plot the scatter plot
ax.scatter(x, y, z, c='r')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Scatter Plot in 3D')

# Show the plot
plt.show()
```

## Surface plot

A surface plot represents a 3D surface or a function in three dimensions. 
It describes a functional relationship between two independent variables X and Z and a designated dependent variable Y, rather than showing the individual data points.

`ax.plot_surface(X, Y, Z)`,

where X and Y are 2D array of points of x and y while Z is 2D array of heights.

When plotting surfaces we need to use the [numpy.meshgrid()](../numpy_and_pandas/numpy_meshgrid.ipynb) function.

### Example: simple surface

```python
x = np.linspace(-2,2,5)
y = np.linspace(-2,2,5)
xx,yy = np.meshgrid(x,y)

zz = xx+yy

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(xx,yy,zz, cmap="Accent")

# Add a color bar which maps values to colors.
fig.colorbar(surface, shrink=0.5)
```

### Example: Rosenbrock function
Let's create a surface plot using the famous [Rosenbrock function](https://en.wikipedia.org/wiki/Rosenbrock_function).

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Generate data for the surface plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface Plot')
```

## Wireframe plot

A wireframe plot visualizes a surface plot with only its wireframe representation. Let's create a wireframe plot of the same Rosenbrock function.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the wireframe
ax.plot_wireframe(X, Y, Z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Wireframe Plot')
```

## Contour plot

A contour plot represents the level curves of a 3D surface in two dimensions. Let's create a contour plot of the Rosenbrock function.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the contour
ax.contour(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Contour Plot')
```

## Surface vs Wireframe vs Contour

```python
x = np.linspace(-2,2,5)
y = np.linspace(-2,2,5)
xx,yy = np.meshgrid(x,y)

zz = xx+yy

fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')
ax3 = fig.add_subplot(133, projection='3d')

ax1.plot_surface(xx,yy,zz, cmap="Accent")
ax2.plot_wireframe(xx,yy,zz, cmap="Accent")
ax3.contour(xx,yy,zz, cmap="Accent")
```
