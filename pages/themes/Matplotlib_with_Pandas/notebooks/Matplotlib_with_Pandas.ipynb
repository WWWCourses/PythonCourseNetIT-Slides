### <a id='toc1_1_4_'></a>[Key Pandas Data Structures](#toc0_)

- Series: One-dimensional labeled arrays.
- DataFrame: Two-dimensional, table-like data structure.
- Series and DataFrames are built on top of NumPy arrays.

#### <a id='toc1_1_4_1_'></a>[Series](#toc0_)

- Series is like a one-dimensional array with labeled data.
- It can hold any data type.
- The labels are called an "index."
- Example: A Series of student names and their corresponding scores.

````python
# Create a dictionary with student names and their scores
data = {
        'Student Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Score': [92, 78, 85, 90, 88]
}

# Create a Pandas Series from the dictionary
student_scores = pd.Series(data['Score'], index=data['Student Name'], dtype=np.int8)

# Display the Series
student_scores
````

#### <a id='toc1_1_4_2_'></a>[DataFrame](#toc0_)
- DataFrame is a two-dimensional, size-mutable, and labeled data structure.
- It's like a spreadsheet or SQL table.
- Rows and columns are both labeled.
- Example: A DataFrame of sales data with rows as transactions and columns as attributes.

````python
# Create a dictionary with student names and their scores
data = {
        'Student Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Score': [92, 78, 85, 90, 88]
}

# Create a Pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
df
````

## <a id='toc1_2_'></a>[Using Matplotlib with Pandas](#toc0_)

One of the most common ways to plot data from a DataFrame is using the `plot` method. This method provides a convenient interface for quickly creating common plots of your DataFrame's data.

Here's an example of how to use the plot method to create a line plot of a DataFrame:

````python
# create a simple DataFrame
df = pd.DataFrame({'x': range(10), 'y': range(10, 20)})
df.head(3)
````

````python
# setup the figure
fig, ax = plt.subplots(figsize=(10, 6), dpi=50)

# plot the data using the 'x' and 'y' columns, specifying the ax
df.plot(x='x', y='y', ax=ax)

# show the plot
plt.show()
````

You can also use the plot method to create other types of plots, such as bar plots, scatter plots, and histograms. For example, here's how you would create a bar plot:

````python
# Variant 1, using 'kind' argument
# df.plot(kind='scatter', x='x', y='y')

# Variant 2, using specific plot method:
df.plot.scatter(x='x', y='y')
plt.show()
````

#### <a id='toc1_2_1_1_'></a>[Plot multiple columns of a DataFrame](#toc0_)

By default, df.plot() will plot all columns in a single axes, if we do not specify the y parameter.

````python
# create a simple DataFrame
df = pd.DataFrame({'x': range(10), 'y1': range(10, 20), 'y2': range(15, 25), 'y3': range(20, 30)})
df.head(3)
````

````python
# Line plot of all columns with 'x' as the x-axis
df.plot(x='x')
plt.title('Line plot of all columns with "x" as x-axis')
plt.xlabel('x')
plt.ylabel('y Values')
plt.show()
````

If we pass to the y parameter a list of column names, it will plot only those columns.

````python
# Line plot y2 and y3 columns with 'x' as the x-axis
df.plot(x='x', y=['y2', 'y3'])
plt.title('Line plot of y1 and y2 columns with "x" as x-axis')
plt.xlabel('x')
plt.ylabel('y Values')
plt.show()
````

#### <a id='toc1_2_1_2_'></a>[Make separate subplots for each column.](#toc0_)

Setting parameter 'subplots=True' matplotlib will make separate subplots for each column.

We can specify which axes the sublots to share, by setting 'sharex' or 'sharey' parameter to True. 

With 'layout=(rows, columns)' we can set the layout of subplots

````python
# Let's have next df:
df = pd.DataFrame({'x': range(10), 'y1': range(10, 20), 'y2': range(15, 25), 'y3': range(20, 30)})
df.head(3)
````

````python
# plot all columns (3rows by 1 column), share X is by default
df.plot(x='x',subplots=True, layout=(3,1))

# plot all columns (1 row by 3 columns), set share Y
df.plot(x='x',subplots=True, sharey=True, layout=(1,3))

plt.show()
````

### <a id='toc1_3_1_'></a>[Best teaching method](#toc0_)

You are given a dataset containing scores achieved by students using different teaching methods ([teaching_methods.csv](https://raw.githubusercontent.com/geekcourses/JupyterNotebooksExamples/master/datasets/various/teaching_methods.csv)). The dataset consists of two columns: Score and Teaching Method. 

Your task is to analyze the dataset to determine which teaching method yields the best results.

#### <a id='toc1_3_1_1_'></a>[Load Dataset in Pandas DataFrame:](#toc0_)

````python
df = pd.read_csv('https://raw.githubusercontent.com/geekcourses/JupyterNotebooksExamples/master/datasets/various/teaching_methods.csv')
df.head(3)
````

#### <a id='toc1_3_1_2_'></a>[Summary statistics](#toc0_)

````python
df.groupby('Teaching Method').describe()
````

Conclusion:

Method 4 has the highest mean score, indicating that it tends to yield higher scores on average compared to the other methods. However, it also has a relatively high standard deviation, indicating some variability in the scores.

Method 3 also shows relatively high scores but has a larger spread (Standard Deviation = 7.51), indicating more variability.

Methods 1 and 2 have lower average scores (Mean Scores around 24.80 and 25.27) and less variability.

#### <a id='toc1_3_1_3_'></a>[Visualizations](#toc0_)

##### <a id='toc1_3_1_3_1_'></a>[Create a box plot to visualize the distribution, median, and outliers of scores for each teaching method](#toc0_)

````python
df.boxplot(column='Score',by='Teaching Method')
plt.xlabel('Teaching Method', color='darkred')
plt.ylabel('Score', color='darkred')
plt.show()
````

##### <a id='toc1_3_1_3_2_'></a>[Create basic histograms for each teaching method to show the distribution of scores.](#toc0_)

````python
df.hist(column='Score',by='Teaching Method', sharex=True, sharey=True, layout=(4,1))
plt.xlabel('Score', color='darkred')
plt.ylabel('Frequency', color='darkred')
plt.show()
````

##### <a id='toc1_3_1_3_3_'></a>[Create histograms with KDE (Kernel Density Estimate) for each teaching method to show the distribution of scores.](#toc0_)

KDE adds a smooth, continuous line to the histogram that represents the estimated density of the data.
Provides a clearer picture of the distribution's shape and central tendency.

To ease the process, we will use [seaborn](https://seaborn.pydata.org/)

````python
import seaborn as sns

plt.figure()

# Create histograms with KDE for each teaching method
methods = df['Teaching Method'].unique()
for method in methods:
    subset = df[df['Teaching Method'] == method]
    sns.histplot(subset['Score'], label=method, kde=True, bins=10, alpha=0.5)

plt.title('Distribution of Scores by Teaching Method')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend(title='Teaching Method')
plt.show()
````

### <a id='toc1_3_2_'></a>[Drinks Consumtion World Wide Stats](#toc0_)

This dataset can be used to analyze alcohol consumption patterns across different countries. Each row represents a country with the corresponding alcohol consumption data, broken down into beer, spirits, wine, and total alcohol consumption in litres. This can be particularly useful for understanding the cultural and regional differences in alcohol consumption habits.

#### <a id='toc1_3_2_1_'></a>[Load DataFrame](#toc0_)

````python
drinks = pd.read_csv('https://raw.githubusercontent.com/geekcourses/JupyterNotebooksExamples/master/datasets/various/drinks.csv')
drinks.head()
````

````python
# get statistical overview of beer consumption per each continent
drinks.groupby('continent')['beer_servings'].describe()
````

#### <a id='toc1_3_2_5_'></a>[Visualizations](#toc0_)

##### <a id='toc1_3_2_5_1_'></a>[Bar Plot of Average drinks consumption by continent](#toc0_)

A bar plot can show the average servings of beer, spirits, and wine, as well as total pure alcohol consumption per continent.

````python
# Calculate the mean values for each continent
drinks_list = ['wine_servings', 'beer_servings', 'spirit_servings' ]
mean_wine = drinks.groupby('continent')[drinks_list].mean()

# Create the bar plot
plt.figure()
mean_wine.plot(kind='bar', edgecolor='black')

# Add title and labels
plt.title('Average Drinks Consumption by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Servings')

plt.show()
````

##### <a id='toc1_3_2_5_2_'></a>[Proportion of each type of alcohol consumption by continent](#toc0_)

Let's create separate pie charts for each type of alcohol (beer_servings, spirit_servings, wine_servings) to show the proportional alcohol consumption by continent.

````python
alcohols_list = ['beer_servings', 'spirit_servings', 'wine_servings']

# Create pie charts for each alcohol type in alcohol list:
fig, axs = plt.subplots(1, len(alcohols_list), figsize=(18, 6))

for idx, alcohol in enumerate(alcohols_list):
    # get sum of alchol consumtion by continent
    alcohol_consumtion = drinks.groupby('continent')[alcohol].sum()
    # create the pie chart
    axs[idx].pie(alcohol_consumtion, labels=alcohol_consumtion.index, autopct='%1.1f%%')
    axs[idx].set_title(alcohol)

# Display the plots
plt.show()
````

##### <a id='toc1_3_2_5_3_'></a>[Plot distribution of wine and beer servings](#toc0_)


The histogram of wine servings shows how many countries fall into each range of wine consumption. A right-skewed distribution might indicate that most countries have low to moderate wine consumption with a few countries having high consumption.

Similarly, the histogram of beer servings will show the spread of beer consumption. Comparing the shape and spread of this distribution with that of wine can reveal differences in how these beverages are consumed globally.

````python
drinks.plot.hist(column= ['beer_servings','wine_servings'],bins=10,alpha = 0.5)
plt.xlabel('Servings')

plt.show()
````

````python
drinks.plot.box(column= ['beer_servings','wine_servings'])
plt.xlabel('Servings')
plt.ylabel('Liters')
plt.show()
````

##### <a id='toc1_3_2_5_4_'></a>[Beer servings vs Wine servings in all continents, accenting on EU and AF](#toc0_)

To compare 2 variables we use scater plot. 
We will plot the dots for EU in blue, the dots for AF in red, for the rest continents - in lightgray.

````python
# lets make a colors array with the shape of drinks, filled with 'lightgray'
colors = np.full_like(drinks.country, 'lightgray')

# assign red color for the EU countries:
colors[drinks.continent=="EU"] = "blue"

# assign lightgreen color for the AF countries:
colors[drinks.continent=="AF"] = "red"

# plot the data
drinks.plot.scatter(x='beer_servings', y='wine_servings',color=colors)

plt.grid(True)
plt.show()
````

Let's see which are those countries which drinks beer in the range [300, 350] and wine in [50,100]

````python
wine_mask = (drinks['wine_servings']>50) & (drinks['wine_servings']<100)
beer_mask = (drinks['beer_servings']>300) & (drinks['beer_servings']<350)

filtered_df = drinks[wine_mask & beer_mask]
filtered_df
````
