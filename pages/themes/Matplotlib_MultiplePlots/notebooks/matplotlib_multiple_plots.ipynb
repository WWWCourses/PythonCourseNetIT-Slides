````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

# Matplotlib: plot multiple figures and plots  

*Note*: Matplotlib quthors discourage working with multiple figures through the implicit pyplot interface because managing the current figure is cumbersome and error-prone. Instead, they recommend using the explicit approach and call methods on Figure and Axes instances.
Reference: [Managing multiple figures in pyplot](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/multiple_figs_demo.html)

````python
# Create some data to be used in next examples:
x = np.linspace(1,100,10)
x
````

## Create multiple axes with figure.add_axes()

The calling signature of add_axes is add_axes(rect), where rect is a list [x0, y0, width, height] denoting the lower left point of the new axes in figure coodinates (x0,y0) and its width and height. So the axes is positionned in absolute coordinates on the canvas

Reference: [add_axes]([https://](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_axes))

````python
fig = plt.figure()

# set axes1 such that it sits tight to the figure boundaries:
axes1 = fig.add_axes([0, 0, 1, 1])

# set axes2 to be at top left corner (leave space for ticks), and size: 1/2 of figure's width and hight:
axes2 = fig.add_axes([0.05, 0.45, 0.5, 0.5]) # inset axes
````

````python
# plot line in axes1
axes1.plot(x, x**2, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# show updated fig
fig
````

````python
# plot line in axes2:
axes2.plot(x,x**3, 'g')
axes2.set_xlabel('x')
axes2.set_ylabel('x**3')
axes2.set_title('insert title')

# show updated fig
fig
````

## Create multiple axes with figure.add_subplot()

`add_subplot()` does not directly provide the option to place the axes at a predefined position. It rather allows to specify where the axes should be situated according to a subplot grid. 
The advantage of this method is that matplotlib takes care of the exact positioning.

The usual and easiest way to specify this position is the 3 integer notation:

`add_subplot(nrows, ncols, index)`

-  The subplot will take the index position on a grid with nrows rows and ncols columns.

Reference: [add_subplot()]([https://](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_subplot))

````python
# Create 2 axes, stacked vertically:
fig = plt.figure()

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')

# adjust layout
plt.tight_layout()
````

Useful parameters are sharex/sharey by which we can make one plot to share the x or y axis of another. The axis will have the same limits, ticks, and scale as the axis of the shared axes.

````python
# Create 2 axes, stacked horizontally
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, sharey=ax1)

# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')

# auto adjust layout, so that there is enough space for labels, ticks and titles
plt.tight_layout()
````

## Create figure and multiple axes with plt.subplots()

`plt.subplots()` is the easiest way to obtain figure and one or more subplots. It returns a tupple of figure object and a single or an array of Axes objects if more than one subplot was created. 

Reference: [plt.subplots()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

````python
# Create 2 axes, stacked vertically:
fig, [ax1,ax2] = plt.subplots(nrows=2, ncols=1)

# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')

# adjust layout
plt.tight_layout()
````

````python
# Create 2 axes, stacked horisontally:
fig, [ax1,ax2] = plt.subplots(nrows=1, ncols=2)

# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')

# adjust layout
plt.tight_layout()
````

We can use again the sharex/sharey parameteres

````python
# Create 2 axes, stacked horisontally:
fig, [ax1,ax2] = plt.subplots(nrows=1, ncols=2, sharey=True)

# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')

# adjust layout
plt.tight_layout()
````

## Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created, using the `figsize` and `dpi` keyword arguments. 

`figsize` is a tuple of the width and height of the figure in **inches**, 
`dpi` is the dots-per-inch (pixel per inch). 

To create an 800x400 pixel figure (100 dots-per-inch), we can do:

````python
fig,ax = plt.subplots(figsize=(8,4), dpi=100)

ax.plot(x, x**2, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
````

Most elements in a figure like lines, markers, texts have a size given in points. Changing the figsize will not scale those elements, but changing the dpi will.

````python
fig,ax = plt.subplots(figsize=(4,2), dpi=200)

ax.plot(x, x**2, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
````

````python
fig,ax = plt.subplots(figsize=(16,8), dpi=50)

ax.plot(x, x**2, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
````

## Saving figures

To save a figure to a file we can use the savefig method in the Figure class.

```
savefig(fname, *, dpi='figure', format=None, metadata=None,
        bbox_inches=None, pad_inches=0.1,
        facecolor='auto', edgecolor='auto',
        backend=None, **kwargs
)	   
```

Note, that we can set the dpi with `dpi` argument

Reference: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html

````python
# Create a figure with two subplots stacked vertically
fig, (ax1, ax2) = plt.subplots(2, 1)

# Plot data
ax1.plot(x, x**2,'r--')
ax2.plot(x, x**3,'g--')

# Set title and labels:
ax1.set_title('Subplot 1')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.set_title('Subplot 2')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.tight_layout()

# Save the figure as png, 640x480:
plt.savefig('figure_640x480.png', dpi=100)

# Save the figure as pdf, as 640x480:
plt.savefig('figure_640x480.pdf',  dpi=100)

# Save the figure as jpg, 1280720:
fig.set_size_inches(12.8,7.2)
plt.savefig('figure_1280720.jpg', dpi=100)

````
