```python
%matplotlib inline 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Ploting various plot types with Matplotlib

## Line

A line chart or line plot is a type of plot that displays data points as connected lines. They are often used to visualize time series data, and are particularly useful for showing trends over time. Line charts can also be used to display continuous data, such as stock prices. Here is an example of a use case for a line chart:

Imagine you are analyzing the temperature of a city over the course of a year. You might want to see how the temperature changes over time. A line chart would be a good choice for this, because it would allow you to see how the temperature changes over the course of the year.

```python
data = pd.read_csv('./datasets/daily-min-temperatures_short.csv')

x = data.Date.values.astype(dtype=np.datetime64)
y = data.Temp.values

plt.plot(x,y)
```

## Scatter plot

A scatter plot is a type of plot that is used to display the relationship between two variables. It uses dots to represent the values of the two variables, and the position of the dots on the plot represents the values of the variables. Scatter plots are often used to investigate the relationship between two continuous variables. 
Here is an example of a use case for a scatter plot:


```python
data = pd.read_csv('./datasets/daily-min-temperatures_short.csv')

x = data.Date.values.astype(dtype=np.datetime64)
y = data.Temp.values

plt.scatter(x,y)
```

## Histogram

Reference: [matplotlib.pyplot.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

A histogram is a type of graph that is used to display the **distribution of a set of continuous or discrete data**. It is particularly useful for showing the distribution of a large number of observations of a single variable. The data is divided into "bins", and the number of observations in each bin is represented by the height of a bar. The bars are adjacent to each other, so it creates a "bar-like" shape, which is why it's called a histogram. 

Consider next example of a use case for a histogram: 
- imagine you are analyzing the test scores of students in a class. You might want to see the distribution of scores across the entire class. A histogram would be a good choice for this, because it would allow you to see how many students scored in each range (e.g., 0-10, 10-20, 20-30, etc.).

To create a histogram, you can use the hist function with the following parameters:

`plt.hist(scores, bins=10`

where:
 - scores: The data we want to plot
 - bins: The number of bins we want to use (10 in this case)


### What is 'bin' ?
In the context of a histogram, a "bin" is a range of values that the data is divided into. Each bin represents a range of values for the data, and the height of each bar in the histogram represents the number of data points that fall within that range of values.

For example, if you have a dataset of test scores for a class and you want to create a histogram to display the distribution of scores, you might choose to use 10 bins. Each bin would represent a range of scores (e.g., 70-79, 80-89, 90-99, etc.), and the height of each bar would represent the number of students who scored within that range.

The choice of bin size can be important because it affects how the data is grouped and represented. If the bin size is too large, the bars may not provide enough detail about the distribution of the data. If the bin size is too small, the histogram may appear too "choppy" and may not convey the overall shape of the distribution. A general rule of thumb is to use a bin size that is the square root of the total number of data points.

```python
# Test scores for the class
scores = np.array([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6],dtype="int")

# Create the histogram
plt.hist(scores, bins=5, edgecolor='black')

# Add labels and title
plt.xlabel('Test Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Test Scores')

# Show the plot
plt.show()
```

If we set the bins parameter to sequence, it defines the bin edges, including the left edge of the first bin and the right edge of the last bin.

In order to see 'bins' on X-axis with values: 2,3,4,5,6 we must set the proper bin edges, i.e. 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5. 

```python
# Test scores for the class
scores = np.array([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6],dtype="int")

# Create the histogram
plt.hist(scores, bins=np.arange(1.5, 7.5, 1), edgecolor='black')

# Add labels and title
plt.xlabel('Test Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Test Scores')

# Show the plot
plt.show()
```

## Bar chart

Reference: [matplotlib.pyplot.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

A bar chart is a chart with rectangular bars with lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. They are used for comparing the values of different categories. Bar charts are often used for displaying categorical data, especially for discrete data with few categories. Here is an example of a use case for a bar chart:

Imagine you are analyzing the sales of different products in a company. You might want to compare the total sales of each product, to see which products are selling the most. A bar chart would be a good choice for this, because it would allow you to see the sales of each product side by side.

```python
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]

plt.bar(products, sales, color = ['red', 'green', 'blue'], edgecolor='black')

plt.show()
```

### Horizontal Bar Chart

Reference: [matplotlib.pyplot.barh](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html#matplotlib.pyplot.barh)

We can create horizontal bar charts with matplotlib.pyplot.barh() method

```python
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]

plt.barh(products, sales, color = ['red', 'green', 'blue'], edgecolor='black')

plt.show()
```
